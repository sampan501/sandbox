```python
from mgc.sims import trans_2samp
from mgc.sims import cubic, uncorrelated_bernoulli
```

```python
import seaborn as sns
import pandas as pd
```

```python
N = 100
P = 3
DEGREE = 90
TRANS = 0
```

```python
def plot_pairplot(sim):
    sim1, sim2 = trans_2samp(sim, N, P, noise=False, degree=DEGREE, trans=TRANS)
    df1 = pd.DataFrame(data=sim1)
    df2 = pd.DataFrame(data=sim2)
    g = sns.pairplot(df1)
    h = sns.pairplot(df2)
```

```python
plot_pairplot(cubic)
```

```python
plot_pairplot(uncorrelated_bernoulli)
```
