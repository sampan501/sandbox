```python
from mgc.sims import trans_2samp
from mgc.sims import cubic, uncorrelated_bernoulli
```

```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
N = 100
P = 3
DEGREE = 90
TRANS = 0
```

```python
def k_sample_transform(inputs):
    n_inputs = len(inputs)
    u = np.vstack(inputs)
    ns = [i.shape[0] for i in inputs]
    v = np.concatenate([np.repeat(i, ns[i]) for i in range(n_inputs)]).reshape(-1, 1)

    return u, v

def plot_pairplot(sim):
    sim1, sim2 = trans_2samp(sim, N, P, noise=False, degree=DEGREE, trans=TRANS)
    u, v = k_sample_transform([sim1, sim2])
    df = pd.DataFrame(data=np.hstack([u, v]))
    df[4].apply(str)
    print(df)
    g = sns.pairplot(df, hue=P+1, vars=list(range(P+1)))
```

```python
plot_pairplot(cubic)
```

```python
plot_pairplot(uncorrelated_bernoulli)
```
